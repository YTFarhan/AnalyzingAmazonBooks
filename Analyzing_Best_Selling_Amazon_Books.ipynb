```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd #importing pandas
import io

df = pd.read_csv(io.BytesIO(uploaded['bestsellers.csv']))
display(df)
```

```python
# These functions will give us a good overview of the data we're working with, including:

# The size of the data
# The column names
# Summary statistics for each column


print(df.head)

print(df.shape)

print(df.columns)

print(df.describe())
```

```python
df.drop_duplicates(inplace=True)  # Remove any duplicate rows in the dataset
```

```python
df.rename(columns={"Name": "Title", "Year": "Publication Year", "User Rating": "Rating"}, inplace=True)

# Rename the columns of the DataFrame
```

```python
df['Price'] = df['Price'].astype(float)  # Convert the "Price" column to a float data type to make it easier to work with
```

```python
author_counts = df['Author'].value_counts() # Which authors have the most books on the Amazon Best Sellers list.
print(author_counts)
```

```python
avg_rating_by_genre = df.groupby('Genre')['Rating'].mean() # What is the average rating for each genre?
print(avg_rating_by_genre)
```

```python
author_counts.head(10).to_csv('top _authors.csv')   # Export top selling authors in csv

avg_rating_by_genre.to_csv('avg_rating_by_genre.csv')  # Export average rating by genre in csv
```
